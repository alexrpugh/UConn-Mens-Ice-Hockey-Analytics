```python
import pandas as pd

# Function to convert time string to total seconds
def time_to_seconds(time_str):
    minutes, seconds = map(int, time_str.split(':'))
    return minutes * 60 + seconds

# Convert 'Team_PP_Time_Elapsed' to total seconds
df['Team_PP_Time_Elapsed_Sec'] = df['Team_PP_Time_Elapsed'].apply(time_to_seconds)
df['Team_PP_OZ possession_Sec'] = df['Team_PP_OZ possession'].apply(time_to_seconds)  
df['Team_PP_NZ possession_Sec'] = df['Team_PP_NZ possession'].apply(time_to_seconds)  
df['Team_PP_DZ possession_Sec'] = df['Team_PP_DZ possession'].apply(time_to_seconds)  

# Calculate the percentages and round to two decimal places
df['Team_PP_OZ possession, %'] = ((df['Team_PP_OZ possession_Sec'] / df['Team_PP_Time_Elapsed_Sec']) * 100).round(2)
df['Team_PP_NZ possession, %'] = ((df['Team_PP_NZ possession_Sec'] / df['Team_PP_Time_Elapsed_Sec']) * 100).round(2)
df['Team_PP_DZ possession, %'] = ((df['Team_PP_DZ possession_Sec'] / df['Team_PP_Time_Elapsed_Sec']) * 100).round(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting histograms for each variable
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
fig.suptitle('Distribution of Power Play Metrics')

# List of columns to plot
cols_to_plot = ['Team_PP_Faceoffs won in OZ, %', 'Team_PP_Faceoffs won, %', 'Team_PP_OZ possession, %', 
                'Team_PP_NZ possession, %', 'Team_PP_DZ possession, %']

for ax, col in zip(axes.flatten(), cols_to_plot):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(col)
    ax.set_xlabel('Percentage')
    ax.set_ylabel('Count')

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Scatter plot for an example pair of variables
# For demonstration, let's choose 'Team_PP_OZ possession, %' and 'Team_PP_Faceoffs won in OZ, %'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Team_PP_OZ possession, %', y='Team_PP_Faceoffs won in OZ, %', data=df, hue='Win')
plt.title('Team PP OZ Possession % vs Faceoffs Won in OZ %')
plt.xlabel('Team PP OZ Possession %')
plt.ylabel('Team PP Faceoffs Won in OZ %')

# Heatmap of correlation matrix
plt.figure(figsize=(10, 6))
correlation_matrix = df[cols_to_plot].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Power Play Metrics')

# Show plots
plt.show()
```
