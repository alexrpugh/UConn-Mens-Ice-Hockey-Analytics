```python
# Function to find identical columns
def find_identical_columns(data, columns):
    identical_columns = []
    checked_pairs = set()
    for i in range(len(columns)):
        for j in range(i+1, len(columns)):
            col1, col2 = columns[i], columns[j]
            if (col1, col2) not in checked_pairs and (col2, col1) not in checked_pairs:
                if data[col1].equals(data[col2]):
                    identical_columns.append((col1, col2))
                checked_pairs.add((col1, col2))
    return identical_columns
```

```python
# Updating the list of column names
column_names_updated = subset_data_dropped.columns.tolist()

# Find identical columns with the updated list of column names
identical_cols_updated = find_identical_columns(subset_data_dropped, column_names_updated)
identical_cols_updated
```

```python
import numpy as np

# Step 3: Handle "-" and "0"
# Converting "-" to NaN
subset_data_cleaned = subset_data_dropped.replace('-', np.nan)

# Checking the first few rows after replacing "-" with NaN
subset_data_cleaned.head()
```

```python
# Step 4: Remove Duplicate Columns

# Helper function to decide which column to keep
def decide_column_to_keep(col1, col2):
    if "Total" in col1 and "PP" in col2:
        return col2
    elif "PP" in col1 and "Total" in col2:
        return col1
    else:
        return col2

# Keeping track of columns to drop
columns_to_drop = set()

# Identifying columns to drop based on the decision criteria
for col1, col2 in identical_cols_updated:
    col_to_drop = decide_column_to_keep(col1, col2)
    columns_to_drop.add(col_to_drop)

# Dropping the identified columns
subset_data_cleaned_dropped = subset_data_cleaned.drop(columns=columns_to_drop)

# Displaying the shape of the DataFrame before and after dropping columns
shape_before = subset_data_cleaned.shape
shape_after = subset_data_cleaned_dropped.shape

shape_before, shape_after, sorted(columns_to_drop)
```

```python
def filter_columns_by_keywords(columns, keywords):
    return [col for col in columns if any(keyword in col.lower() for keyword in keywords)]

# Grouping columns by categories
column_categories = {
    "Basic Game Information": ['Date', 'Opponent', 'Home', 'Win'],
    "Custom Calculated Columns": ['Team_Score_On_PP', 'Opponent_Score_On_PP'],
}

# All other categories will be filled based on keywords
keywords_categories = {
    "Team Power Play Statistics": ['team_pp_'],
    "Opponent Power Play Statistics": ['opponent_pp_'],
    "Team Total Statistics": ['team_total_'],
    "Opponent Total Statistics": ['opponent_total_'],
    "Faceoff Statistics": ['faceoff'],
    "Shot Statistics": ['shot'],
    "Possession Statistics": ['possession'],
    "Pass and Play Statistics": ['pass', 'play', 'entry', 'breakout'],
    "Penalty and Retrieval Statistics": ['penalty', 'retrieval'],
    "Scoring and Goals": ['score', 'goal', 'shootout'],
    "Penalties and Hits": ['penalt', 'hit'],
    "Puck Battles and Possession": ['puck battle', 'deke', 'puck loss', 'loose puck recovery'],
    "Entries and Counterattacks": ['entry', 'dump in', 'counterattack'],
    "Scoring Chances and Conversion": ['scoring chance', 'corsi%', 'xg'],
    "Location-Based Actions": ['oz', 'nz', 'dz'],
    "Hits and Takeaways": ['hit', 'takeaway']
}

# Populate categories based on keywords
for category, keywords in keywords_categories.items():
    column_categories[category] = filter_columns_by_keywords(subset_data_cleaned.columns, keywords)

# Identify miscellaneous columns (those not categorized yet)
all_categorized_columns = set(col for cols in column_categories.values() for col in cols)
miscellaneous_columns = set(subset_data_cleaned.columns) - all_categorized_columns
column_categories["Miscellaneous Statistics"] = list(miscellaneous_columns)

# Displaying the number of columns in each category
for category, columns in column_categories.items():
    print(f"{category}: {len(columns)} columns")
```
