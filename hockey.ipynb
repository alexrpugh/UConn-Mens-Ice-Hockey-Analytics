```python
import pandas as pd

total_stats = pd.read_excel('/Users/giovanni-lunetta/uconn_masters/hockey/hockey_repo/data/Total Stats.xlsx')
pp_stats = pd.read_excel('/Users/giovanni-lunetta/uconn_masters/hockey/hockey_repo/data/PP Stats.xlsx')
```

```python
total_stats = total_stats.iloc[:-2]
pp_stats = pp_stats.iloc[:-2]
```

```python
total_stats.shape
```

```python
pp_stats.shape
```

```python
# 2. Separate the rows based on the "Teams" column
team_stats = total_stats[total_stats['Teams'] == 'UH'].reset_index(drop=True)
opponent_stats = total_stats[total_stats['Teams'] != 'UH'].reset_index(drop=True)

# Drop the 'Teams', 'Date', and 'Opponent' columns as they will be redundant after the merge
team_stats = team_stats.drop(columns=['Teams', 'Date', 'Opponent'])
opponent_stats = opponent_stats.drop(columns=['Teams', 'Date', 'Opponent'])

# 3. Add appropriate prefixes to column names
team_stats.columns = ["Team_Total_" + col for col in team_stats.columns]
opponent_stats.columns = ["Opponent_Total_" + col for col in opponent_stats.columns]

# 4. Merge the datasets side by side
combined_total_stats = pd.concat([team_stats, opponent_stats], axis=1)

# Add the 'Date' and 'Opponent' columns back to the combined dataset
combined_total_stats['Date'] = total_stats['Date'][total_stats['Teams'] == 'UH'].reset_index(drop=True)
combined_total_stats['Opponent'] = total_stats['Opponent'][total_stats['Teams'] == 'UH'].reset_index(drop=True)

# Reorder columns to have 'Date' and 'Opponent' at the beginning
column_order_total = ['Date', 'Opponent'] + [col for col in combined_total_stats if col not in ['Date', 'Opponent']]
combined_total_stats = combined_total_stats[column_order_total]
```

```python
combined_total_stats
```

```python
# 1. Separate the rows based on the "Teams" column for "PP Stats" dataset
team_pp_stats = pp_stats[pp_stats['Teams'] == 'UH'].reset_index(drop=True)
opponent_pp_stats = pp_stats[pp_stats['Teams'] != 'UH'].reset_index(drop=True)

# Drop the 'Teams', 'Date', and 'Opponent' columns as they will be redundant after the merge
team_pp_stats = team_pp_stats.drop(columns=['Teams', 'Date', 'Opponent'])
opponent_pp_stats = opponent_pp_stats.drop(columns=['Teams', 'Date', 'Opponent'])

# 2. Add appropriate prefixes to column names
team_pp_stats.columns = ["Team_PP_" + col for col in team_pp_stats.columns]
opponent_pp_stats.columns = ["Opponent_PP_" + col for col in opponent_pp_stats.columns]

# 3. Merge the datasets side by side
combined_pp_stats = pd.concat([team_pp_stats, opponent_pp_stats], axis=1)

# Add the 'Date' and 'Opponent' columns back to the combined dataset
combined_pp_stats['Date'] = pp_stats['Date'][pp_stats['Teams'] == 'UH'].reset_index(drop=True)
combined_pp_stats['Opponent'] = pp_stats['Opponent'][pp_stats['Teams'] == 'UH'].reset_index(drop=True)

# Reorder columns to have 'Date' and 'Opponent' at the beginning
column_order_pp = ['Date', 'Opponent'] + [col for col in combined_pp_stats if col not in ['Date', 'Opponent']]
combined_pp_stats = combined_pp_stats[column_order_pp]
```

```python
combined_pp_stats
```

```python
combined_stats = pd.merge(combined_total_stats, combined_pp_stats, on=['Date', 'Opponent'])
```

```python
combined_stats
```

```python
combined_stats['Home'] = combined_stats['Opponent'].str.startswith('vs')
```

```python
# Remove "vs " and "@ " from the 'Opponent' column
combined_stats['Opponent'] = combined_stats['Opponent'].str.replace('vs ', '').str.replace('@ ', '')
```

```python
# Create a 'UConn_Won' column based on the comparison of scores
combined_stats['Win'] = combined_stats['Team_Total_UConn Score'] > combined_stats['Team_Total_Opponent Score']
```

```python
# Remove the specified columns
combined_stats.drop(columns=['Team_Total_Goals', 'Team_PP_Goals'], inplace=True)
```
